```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Prepare features and target
X = df[['year']]          # Independent variable
y = df['per capita income (US$)']  # Dependent variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict per capita income for year 2020
prediction_2020 = model.predict([[2020]])

print("Predicted per capita income for Canada in 2020:", prediction_2020[0])
```
